```python
import matplotlib.pyplot as plt
import numpy as np
%config InlineBackend.figure_formats = ['svg']

data = np.load('../data/obstacles.npz')
```

```python
color_images = data['color_images']
depth_images = data['depth_images']
maps = data['maps']
cmd_vels = data['cmd_vels']
```

```python
print(color_images.shape)
print(depth_images.shape)
print(maps.shape)
print(cmd_vels.shape)
```

```python
plt.figure(figsize=(14,4))
plt.plot(cmd_vels[:,0], label='u_x')
plt.plot(cmd_vels[:,1], label='u_y')
plt.grid()
plt.legend()
plt.show()
```

```python
indices = range(4000,4300)

plt.figure(figsize=(6,4))
plt.plot(cmd_vels[indices,0], label='u_x')
plt.plot(cmd_vels[indices,1], label='u_y')
plt.xlabel('Timesteps')
plt.ylabel('Control Input')
plt.grid()
plt.legend()
plt.show()
```
