```python
df2_used['field_vel']=None
```

```python
for i, time in enumerate(df2_used['%time']):
    #print(i)
    #print(time)
    result_index = df1_used['%time'].sub(time).abs().idxmin()
    df2_used['field_vel'][i:i+1] = df1_used['field_vel'][result_index:result_index+1]
```

```python
import re
import ast
import torch

grids = torch.as_tensor([[int(num) for num in re.findall(r'-?\d+', string)] for string in df2_used["field.data"].to_list()])
commands = torch.as_tensor([ast.literal_eval(string) for string in df2_used["field_vel"].to_list()])

# reshape and turn 100s into 1s
grids = grids.view((grids.shape[0], 100, 100))
grids[grids==100]=1
```

```python
torch.save({grids, commands}, 'data/robot_field_data.pt')
```
